`The need for normalization becomes clear when we look at the range and scale differences.`

Evidence is as follows:
Feature: radius_mean
- Min: 6.98, Max: 28.11

Feature: area_mean
- Min: 143.5, Max: 2501.0

Feature: smoothness_mean
- Min: 0.05263, Max: 0.16340

Why it matter?
- area_mean is in hundreds to thousands,
- smoothness_mean is between 0.05 and 0.16,
- radius_mean is in tens.

These drastically different scales (0.05 → 2501) will cause Logistic Regression and optimization algorithms to behave poorly because:
- Larger-scale features dominate the cost function.
- Gradient descent convergence becomes inefficient.

Clear Indicator shows:
area_mean → 2501
smoothness_mean → 0.16
This difference (a factor of 15,000+) clearly signals the need for normalization.
