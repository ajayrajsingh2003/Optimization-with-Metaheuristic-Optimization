# Objective:

Prepare the dataset for modeling by:
- Dropping irrelevant columns.
- Encoding target variable.
- Splitting features and target.
- Applying normalization using MinMaxScaler.

Why Scaling?
Feature scales vary drastically (e.g., `area_mean` up to 2501 vs. `smoothness_mean` ~0.16).
Without normalization,
- large-scale features dominate the model, affecting performance.

`Import Libraries`

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

`Load Dataset`

```python
df = pd.read_csv('C:/Users/ajayr/Desktop/Projects to upload/Metaheuristic_Optimization_for_Logistic_Regression/data/data.csv')
```

```python
df
```

`Drop Irrelevant Columns`

```python
# 'id' column is not useful for prediction, drop it
if 'id' in df.columns:
    df.drop(columns=['id'], inplace=True)
```

```python
df
```

`Encode Target Variable`

```python
# Convert 'diagnosis' (M = malignant, B = benign) to numeric
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Check conversion
df['diagnosis'].value_counts()
```

`Split Features (X) and Target (y)`

```python
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

`Apply MinMax Scaling`

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for readability
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Check scaled feature ranges
X_scaled.describe().T
```

# Summary

- Removed irrelevant columns (`id`, unnamed).
- Encoded target variable (Malignant=1, Benign=0).
- Scaled all features using MinMaxScaler.
- Data is ready for model building in the next notebook.
